#Q2. Explain the difference between Absolute and Relative positioning.

Ans. 1. Absolute Positioning (position: absolute;)
The element is removed from the normal document flow.

It is positioned relative to the nearest positioned ancestor (i.e., an ancestor with position: relative, absolute, or fixed).

If there is no positioned ancestor, it is positioned relative to the <html> (or the nearest containing block).

It does not affect surrounding elements.

2. Relative Positioning (position: relative;)
The element remains in the normal document flow.

It is positioned relative to its normal position.

Other elements are not affected by its movement.

If you donâ€™t specify top, left, right, or bottom, it remains in its original position.

#Q.2 Explain the z-index, with a code example.

The z-index property in CSS is used to control the stacking order of elements. Elements with a higher z-index value appear in front of elements with a lower z-index value.

Works only on elements with a position value of relative, absolute, or fixed (not static).

The default z-index is auto, meaning it follows the natural stacking order.

Higher z-index values bring elements to the front.

If two elements have the same z-index, the later one in the HTML appears in front.
